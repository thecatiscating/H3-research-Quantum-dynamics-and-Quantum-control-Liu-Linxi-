```python
import numpy as np
import matplotlib.pyplot as plt
from qutip import *

# Define the Bloch vectors for the lab frame and rotating frame
lab_frame_vector = [1, 0, 0]
rotating_frame_vector = [np.cos(np.pi), np.sin(np.pi), 0]  # Replace theta with your angle

# Create Bloch spheres
bloch_lab = Bloch()
bloch_lab.add_vectors(lab_frame_vector)
bloch_lab.show()

bloch_rotating = Bloch()
bloch_rotating.add_vectors(rotating_frame_vector)
bloch_rotating.show()
```
